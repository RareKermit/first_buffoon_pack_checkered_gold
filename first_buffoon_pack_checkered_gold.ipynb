Now that we have an approximate evaluation for all the jokers on Checkered Deck, let's write some functions.

```python
import random

def roll_rarity():
    """
    Returns 'common', 'uncommon', or 'rare'.
    
    Mirrors Balatro's rarity poll from get_current_pool in common_events.lua:
      - rarity > 0.95 -> rare     (5%)
      - rarity > 0.7  -> uncommon (25%)
      - else          -> common   (70%)
    """
    r = random.random()
    if r > 0.95:
        return "rare"
    if r > 0.7:
        return "uncommon"
    return "common"

```

```python
outcomes = []

for _ in range(100_000):
    outcomes.append(roll_rarity())

print("commons",outcomes.count('common'))
print("uncommons",outcomes.count('uncommon'))
print("rares",outcomes.count('rare')) 
```

The function roll_rarity is kinda straightforward, but the stickers were not, and required a bit more investigation of the game's code.
Turns out there's basically 2 dice involved.  You can check out the function below, etc.

```python
def roll_stickers():
    """
    Returns a tuple (eternal, perishable, rental), where each is either "sticker_name" or "" for each category
    
    Mirrors Balatro's gold-stake sticker logic from common_events.lua:
    
    if _type == 'Joker' then
        if G.GAME.modifiers.all_eternal then
            card:set_eternal(true)
        end
        if (area == G.shop_jokers) or (area == G.pack_cards) then 
            local eternal_perishable_poll = pseudorandom((area == G.pack_cards and 'packetper' or 'etperpoll')..G.GAME.round_resets.ante)
            if G.GAME.modifiers.enable_eternals_in_shop and eternal_perishable_poll > 0.7 then
                card:set_eternal(true)
            elseif G.GAME.modifiers.enable_perishables_in_shop and ((eternal_perishable_poll > 0.4) and (eternal_perishable_poll <= 0.7)) then
                card:set_perishable(true)
            end
            if G.GAME.modifiers.enable_rentals_in_shop and pseudorandom((area == G.pack_cards and 'packssjr' or 'ssjr')..G.GAME.round_resets.ante) > 0.7 then
                card:set_rental(true)
            end
        end

    TLDR:
    Dice 1: eternal 30%, perishable 30%, neither 40%
    Dice 2: rental 30%, not rental 70%
    """
    dice1 = random.random()
    if dice1 > 0.7:
        eternal, perishable = "eternal", ""
    elif dice1 > 0.4:
        eternal, perishable = "", "perishable"
    else:
        eternal, perishable = "", ""

    dice2 = random.random()
    
    rental = "rental" if dice2 < 0.3 else ""
    
    return eternal, perishable, rental
```

Now we make sure the function is working.

```python
results = {}
results["eternal_exists"] = 0
results["perishable_exists"] = 0
results["rental_exists"] = 0
results["clean"] = 0
results["eternal"] = 0
results["perishable"] = 0
results["rental"] = 0
results["eternal_rental"] = 0
results["perishable_rental"] = 0

for _ in range(100_000):
    tup = roll_stickers()
    eternal_status = tup[0]
    perishable_status = tup[1]
    rental_status = tup[2]
    if eternal_status:
        results["eternal_exists"] += 1
    if perishable_status:
        results["perishable_exists"] += 1
    if rental_status:
        results["rental_exists"] += 1
    if eternal_status and rental_status:
        results["eternal_rental"] += 1
    if perishable_status and rental_status:
        results["perishable_rental"] += 1
    if eternal_status and not perishable_status and not rental_status:
        results["eternal"] += 1
    if not eternal_status and perishable_status and not rental_status:
        results["perishable"] += 1
    if not eternal_status and not perishable_status and rental_status:
        results["rental"] += 1
    if not eternal_status and not perishable_status and not rental_status:
        results["clean"] += 1
    if perishable_status and eternal_status:
        print("burp")

print(results)
```

Now we can generate a random joker with random stickers.

```python
def generate_joker(jokers_dictionary,rarity_override=False,name_override=False):
    """
    Pick a random joker from the dictionary and roll its stickers, honoring compatibility.
    
    Returns (rarity, name, sticker_key) where sticker_key can be 'clean', 'eternal', 'perishable', 'rental',
    'eternal_rental', 'perishable_rental'
    """
    if rarity_override == False:
        rarity = roll_rarity()
    else:
        rarity = rarity_override
    if name_override == False:
        name = random.choice(list(jokers_dictionary[rarity].keys()))
    else:
        name = name_override
        
    data = jokers_dictionary[rarity][name]
    
    eternal, perishable, rental = roll_stickers()
    
    # if the joker has "" for that sticker in the dictionary, it can't have it.
    if eternal == "eternal" and data.get("eternal") == "":
        eternal = ""
    if perishable == "perishable" and data.get("perishable") == "":
        perishable = ""
    #doesn't exist yet but could in the future who knows
    if rental == "rental" and data.get("rental") == "":
        rental = ""
    
    if eternal and rental:
        sticker_key = "eternal_rental"
    elif perishable and rental:
        sticker_key = "perishable_rental"
    elif eternal:
        sticker_key = "eternal"
    elif perishable:
        sticker_key = "perishable"
    elif rental:
        sticker_key = "rental"
    else:
        sticker_key = "clean"
    
    return rarity, name, sticker_key
```

Alright, now we just gotta make a pack, which is just making sure the joker name isn't the same / regenerating at the same rarity.

```python
def generate_pack(jokers_dictionary):
    joker_one = generate_joker(jokers_dictionary)
    joker_two = generate_joker(jokers_dictionary)
    while joker_one[1] == joker_two[1]: #while names are equal (should never happen)
        joker_two = generate_joker(jokers_dictionary,rarity_override=joker_two[0])

    return joker_one,joker_two
```

Last function, just pick what is best from the 2 options available.

```python
def pick_best(jonklers,jokers_dictionary):
    best_so_far = 0
    
    first_choice = jokers_dictionary[jonklers[0][0]][jonklers[0][1]][jonklers[0][2]]
    #print(first_choice)
    if first_choice == "good":
        best_so_far = "good"
    elif type(best_so_far) == int:
        if best_so_far < first_choice:
            best_so_far = first_choice
    second_choice = jokers_dictionary[jonklers[1][0]][jonklers[1][1]][jonklers[1][2]]
    #print(second_choice)
    if second_choice == "good":
        best_so_far = "good"
    elif type(best_so_far) == int:
        if best_so_far < second_choice:
            best_so_far = second_choice
    return best_so_far

```

```python
test_jonklers = generate_pack(jokers)
print(test_jonklers)

best = pick_best(test_jonklers,jokers)
print("choice",best)
```

```python
from collections import Counter

N = 1_000_000
outcomes = Counter()
for _ in range(N):
    j1, j2 = generate_pack(jokers)
    outcomes[pick_best((j1, j2), jokers)] += 1

# Print results
for k, v in sorted(outcomes.items(), key=lambda x: (-1 if x[0] == 'good' else x[0])):
    label = 'good' if k == 'good' else f'${k}'
    print(f"  {label:>8}: {v:>16} ({100*v/N:5.2f}%)")
```
